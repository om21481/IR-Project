```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Function to extract tables from a URL and save them to a text file
def extract_tables_and_save(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        tables = soup.find_all('table')
        with open(filename, 'w') as f:
            for table in tables:
                rows = table.find_all('tr')
                for row in rows:
                    cells = row.find_all(['th', 'td'])
                    row_data = [cell.get_text(strip=True) for cell in cells]
                    row_text = '\t'.join(row_data)
                    f.write(row_text + '\n')
                f.write('\n' + '-' * 50 + '\n\n')
        print(f"Tables extracted successfully and saved to '{filename}'")
    else:
        print(f"Failed to retrieve data for {url}")

# Read the CSV file
csv_filename = 'ind_nifty50list.csv'  # Replace 'your_csv_file.csv' with the actual filename
data = pd.read_csv(csv_filename)

# Ensure a directory to save the text files exists
output_directory = 'extracted_tables'
os.makedirs(output_directory, exist_ok=True)

# Iterate over each row in the Symbol column
for index, row in data.iterrows():
    symbol = row['Symbol']
    url = f'https://chartink.com/fundamentals/{symbol}.html'
    output_filename = os.path.join(output_directory, f'{symbol}_tables.txt')
    extract_tables_and_save(url, output_filename)
```

```python
import os

# Function to minimize a text file
def minimize_text_file(input_filename, output_filename):
    with open(input_filename, 'r') as input_file:
        lines = input_file.readlines()

    minimized_lines = []
    for line in lines:
        # Skip empty lines
        if line.strip():
            # Remove leading and trailing whitespace
            minimized_lines.append(line.strip())

    with open(output_filename, 'w') as output_file:
        output_file.write('\n'.join(minimized_lines))

    print(f"File '{input_filename}' minimized and saved as '{output_filename}'")

# Folder containing text files
folder_path = 'C:\\Users\\SIDDHARTH\\Desktop\\ir project\\extracted_tables'  # Replace 'your_folder_path' with the actual folder path
output_folder = 'C:\\Users\\SIDDHARTH\\Desktop\\ir project\\minimised_tables'
os.makedirs(output_folder, exist_ok=True)

# Iterate through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        input_file_path = os.path.join(folder_path, filename)
        output_file_path = os.path.join(output_folder, filename)
        minimize_text_file(input_file_path, output_file_path)
```
